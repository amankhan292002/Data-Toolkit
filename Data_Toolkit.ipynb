#Data Toolkit


1.  What is NumPy, and why is it widely used in Python?
    - NumPy is a Python library used for numerical computing. It provides fast and efficient array operations, mathematical functions, and support for large datasets, making it widely used in data science and scientific computing.


2.  How does broadcasting work in NumPy?
     - Broadcasting allows NumPy to perform operations on arrays of different shapes by automatically expanding the smaller array to match the larger one without copying data.

3. What is a Pandas DataFrame?
    - A Pandas DataFrame is a two-dimensional labeled data structure with rows and columns. It is used to store and manipulate structured data like tables or spreadsheets.

4. Explain the use of the groupby() method in Pandas.
    - The groupby() method is used to group data based on one or more columns and then apply aggregation functions such as sum, mean, or count.

5.  Why is Seaborn preferred for statistical visualizations?
    - Seaborn provides built-in themes and high-level functions that make it easier to create attractive and informative statistical plots with less code.

6.  What are the differences between NumPy arrays and Python lists?
    - NumPy arrays are faster, consume less memory, and support vectorized operations, while Python lists are more flexible and can store different data types.

7.  What is a heatmap, and when should it be used?
    - A heatmap is a data visualization technique that uses colors to represent values in a matrix. It is useful for identifying patterns, correlations, or intensity in large datasets.

8. What does the term “vectorized operation” mean in NumPy?
    - A vectorized operation allows calculations to be applied to entire arrays at once without using explicit loops, resulting in faster execution.

9. How does Matplotlib differ from Plotly?
    - Matplotlib is mainly used for static plots, while Plotly is used for interactive and dynamic visualizations, especially in web applications.

10.  What is the significance of hierarchical indexing in Pandas?
     - Hierarchical indexing allows multiple index levels on rows or columns, enabling the handling of higher-dimensional data in a compact and organized way.

11. What is the role of Seaborn's pairplot() function?
     - The pairplot() function creates pairwise plots for numerical variables, helping visualize relationships and distributions in a dataset.

12. What is the purpose of the describe() function in Pandas?
     - The describe() function provides summary statistics such as count, mean, standard deviation, minimum, and maximum for numerical data.

13. Why is handling missing data important in Pandas?
     - Handling missing data is important to ensure accurate analysis, prevent errors in computations, and maintain data quality.

14. What are the benefits of using Plotly for data visualization?
    - Plotly offers interactive charts, zooming, tooltips, and easy integration with web applications, making it useful for dashboards and presentations.

15. How does NumPy handle multidimensional arrays?
    - NumPy supports multidimensional arrays using the ndarray object, allowing efficient storage and operations on data with multiple dimensions.

16. What is the role of Bokeh in data visualization?
     - Bokeh is used to create interactive and web-based visualizations that can handle large datasets and real-time updates.

17.  Explain the difference between apply() and map() in Pandas.
     - apply() works on rows or columns of a DataFrame, while map() is used on a Series to transform individual values.

18. What are some advanced features of NumPy?
    - Advanced features include broadcasting, vectorization, linear algebra operations, random number generation, and integration with C/C++ code.

19. How does Pandas simplify time series analysis?
     - Pandas provides date-time indexing, resampling, rolling windows, and built-in time series functions for easy analysis.

20. What is the role of a pivot table in Pandas?
    - A pivot table summarizes and reorganizes data by grouping and aggregating values based on selected columns.

21. Why is NumPy's array slicing faster than Python's list slicing?
     - NumPy arrays use contiguous memory and optimized C-level operations, making slicing faster than Python lists.


22. What are some common use cases for Seaborn?
    - Seaborn is commonly used for statistical analysis, data exploration, correlation analysis, and visualizing distributions and relationships.

```python
# 1.How do you create a 2D NumPy array and calculate the sum of each row?

import numpy as np

arr = np.array([[1, 2, 3],
                [4, 5, 6]])

row_sum = arr.sum(axis=1)
print("Sum of each row:", row_sum)
```

```python
# 2. Write a Pandas script to find the mean of a specific column in a DataFrame

import pandas as pd

df = pd.DataFrame({
    "Marks": [70, 80, 90, 85]
})

print("Mean:", df["Marks"].mean())
```

```python
# 3.  Create a scatter plot using Matplotlib

import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 15, 25, 30]

plt.scatter(x, y)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
# 4.  How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "A": [1, 2, 3, 4],
    "B": [4, 3, 2, 1],
    "C": [2, 3, 4, 5]
})

corr = df.corr()
sns.heatmap(corr, annot=True)
plt.show()
```

```python
# 5.  Generate a bar plot using Plotly

import plotly.express as px

data = {"Fruits": ["Apple", "Banana", "Orange"],
        "Count": [10, 15, 7]}

fig = px.bar(data, x="Fruits", y="Count")
fig.show()
```

```python
# 6. Create a DataFrame and add a new column based on an existing column.
import pandas as pd

df = pd.DataFrame({
    "Price": [100, 200, 300]
})

df["Discounted_Price"] = df["Price"] * 0.9
print(df)
```

```python
# 7. Write a program to perform element-wise multiplication of two NumPy arrays

import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

result = a * b
print("Element-wise multiplication:", result)
```

```python
# 8. Create a line plot with multiple lines using Matplotlib.

import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [10, 20, 30, 40]
y2 = [15, 25, 35, 45]

plt.plot(x, y1, label="Line 1")
plt.plot(x, y2, label="Line 2")
plt.legend()
plt.show()
```

```python
# 9. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.

import pandas as pd

df = pd.DataFrame({
    "Score": [50, 70, 90, 65]
})

filtered = df[df["Score"] > 70]
print(filtered)
```

```python
# 10. Create a histogram using Seaborn to visualize a distribution.

import seaborn as sns
import matplotlib.pyplot as plt

data = [10, 20, 20, 30, 30, 30, 40]

sns.histplot(data, bins=5)
plt.show()
```

```python
# 11. Perform matrix multiplication using NumPy.

import numpy as np

a = np.array([[1, 2],
              [3, 4]])

b = np.array([[5, 6],
              [7, 8]])

result = np.dot(a, b)
print("Matrix multiplication result:\n", result)
```

```python
# 13. Create a 3D scatter plot using Plotly.

import plotly.express as px
import pandas as pd

df = pd.DataFrame({
    "x": [1, 2, 3, 4],
    "y": [10, 20, 30, 40],
    "z": [5, 15, 25, 35]
})

fig = px.scatter_3d(df, x="x", y="y", z="z")
fig.show()
```
